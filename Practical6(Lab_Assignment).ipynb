```python
# 1.This script prompts the user for a string and then iterates through it to count specific categories of characters
def get_string_statistics():
    user_input = input("Enter a string: ")

    # Initialize counters
    vowels_list = "aeiouAEIOU"
    vowels = 0
    consonants = 0
    spaces = 0
    lowercase = 0

    for char in user_input:
        # Check for vowels and consonants
        if char.isalpha():
            if char in vowels_list:
                vowels += 1
            else:
                consonants += 1

        # Check for spaces
        if char.isspace():
            spaces += 1

        # Check for lowercase letters
        if char.islower():
            lowercase += 1

    # Print results
    print(f"\nStatistics for: '{user_input}'")
    print(f"a) Number of Vowels: {vowels}")
    print(f"b) Number of Consonants: {consonants}")
    print(f"c) Number of Spaces: {spaces}")
    print(f"d) Number of Lowercase Letters: {lowercase}")

if __name__ == "__main__":
    get_string_statistics()
```

```python
#2. This function is designed to handle multiple lines of input (a sequence) and convert them all to uppercase
def capitalize_sequence():
    print("Enter your lines (press Enter on an empty line to finish):")
    lines = []

    while True:
        line = input()
        if line == "":
            break
        lines.append(line)

    print("\nOutput:")
    for line in lines:
        print(line.upper())

if __name__ == "__main__":
    capitalize_sequence()
```
