```python
#1. This program focuses on basic tuple operations

# Step 1: Input a series of integers and store them in a Tuple
user_input = input("Enter a series of integers separated by spaces: ")
# Convert the input string into a tuple of integers
my_tuple = tuple(int(x) for x in user_input.split())

# a) Print the total number of items in the Tuple [cite: 8]
print(f"Total number of items: {len(my_tuple)}")

# b) Print the last item in the Tuple [cite: 9]
print(f"Last item: {my_tuple[-1]}")

# c) Print the Tuple elements in reverse order [cite: 10]
print(f"Tuple in reverse: {my_tuple[::-1]}")

# d) Print Yes if the Tuple contains an integer 5 and No otherwise [cite: 11]
if 5 in my_tuple:
    print("Yes")
else:
    print("No")

# e) Remove the first and last items, sort the rest, and print [cite: 12]
remaining_items = list(my_tuple[1:-1])
remaining_items.sort()
print(f"Remaining sorted items: {remaining_items}")
```

```python
#2. This program manages sales data
# Store the Prices of sold items in a Tuple [cite: 14]
# Example input: 50 120 45 200 200 80
prices_input = input("Enter the prices of items sold today (separated by spaces): ")
prices = tuple(float(x) for x in prices_input.split())

# a) Print the total number of items sold [cite: 16]
print(f"Total items sold: {len(prices)}")

# b) Print the price of cheapest item sold [cite: 17]
print(f"Cheapest item: {min(prices)}")

# c) Print the price of costliest item sold [cite: 18]
costliest = max(prices)
print(f"Costliest item: {costliest}")

# d) Print the price list in ascending order [cite: 19]
print(f"Prices in ascending order: {sorted(prices)}")

# e) Print the number of costliest items sold on the day [cite: 20]
count_costliest = prices.count(costliest)
print(f"Number of costliest items sold: {count_costliest}")
```
