```python
#1.Print numbers from 1 to 10

for i in range(1, 11):
    print(i)
```

```python
#2.Print numbers from 10 to 1

for i in range(10, 0, -1):
    print(i)
```

```python
#3.Print numbers from 1 to n without spaces

n = int(input("Enter a number: "))

for i in range(1, n + 1):
    print(i, end="")
```

```python
#4. Print the star pattern

rows = 5

for i in range(rows):
    print(" " * i, end="")

    for j in range(rows - i):
        print("*", end=" ")

    print()
```

```python
# 5. Print the star pattern

rows = 5

for i in range(1, rows + 1):

    for j in range(i):
        print("*", end=" ")

    print()
```

```python
# 6. Print the number pattern

rows = 5

for i in range(1, rows + 1):

    for j in range(1, i + 1):
        print(j, end=" ")

    print()
```

```python
#7.To find all prime numbers between two numbers
while True:
    start = int(input("Enter starting number: "))
    end = int(input("Enter ending number: "))

    if start < end:
        break
    else:
        print("Starting number must be smaller than ending number. Try again.")

print("Prime numbers between", start, "and", end, "are:")

for num in range(start, end + 1):

    if num > 1:
        prime = True

        for i in range(2, num):
            if num % i == 0:
                prime = False
                break

        if prime:
            print(num)
```
